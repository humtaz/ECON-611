```python
# 1. Generate data for one random variable with a mean of 5 and standard deviation of 1. 
# Run a one-sample t-test that the mean of the sample is different than 0.
# Perform the same test again with a mean of 1 and a standard deviation of 1.
```

```python
import numpy as np
from scipy import stats
np.random.seed(100)

# mean = 5, std = 1;
mean_5 = np.random.normal(loc=5, scale=1, size=100)
t_test_5 = stats.ttest_1samp(mean_5, 0)
print("Results for mean = 5 and SD = 1:")
print("T-test:", round(t_test_5.statistic, 3))


# mean = 1, std = 1
mean_1 = np.random.normal(loc=1, scale=1, size=100)
t_test_1 = stats.ttest_1samp(mean_1, 0)
print("\nResults for data with mean = 1 and SD = 1:")
print("T-test:", round(t_test_1.statistic, 3))
```

```python
# 2. Why has the result changed? In answering this question, plot both samples with the index as the x-axis values. 
# Use two colors to differentiate the samples. Indicate the value of zero using a horizontal line. Reference the plot in your answer.
```

The result of the one-sample t-test has changed because the underlying data distributions have different means. 
In the first case, the data is generated with a mean of 5, while in the second case, the data is generated with a mean of 1.
When we perform a one-sample t-test, we compare the mean of our sample data to a hypothesized population mean (in this case, 0). If the mean of our sample significantly differs from the hypothesized mean, we reject the null hypothesis.
Plotting both samples will provide visual insight into their distributions and how they differ from the hypothesized mean of 0. 

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.scatter(range(len(mean_5)), mean_5, color='blue', label='Mean = 5')
plt.scatter(range(len(mean_1)), mean_1, color='red', label='Mean = 1')

# Plot horizontal line at zero
plt.axhline(y=0, color='black', linestyle='--', label='Zero')

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Comparison of Samples')
plt.legend()
plt.grid(True)
plt.show()
```
