```python
# 1. Generate a sample of 100 observations drawn from a normal distribution with mean of 2 and standard deviation of .5. Plot the result using a histogram.


import numpy as np
import matplotlib.pyplot as plt

mean = 2
std_dev = 0.5
sample_size = 100
data = np.random.normal(mean, std_dev, sample_size)

plt.hist(data, bins=20, density=True, alpha=0.6, color='b')
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 2. Create a cumulative sum from the values created in the previous exercise. Plot the result. Does the data appear to follow a trend? Explain.

mean = 2
std_dev = 0.5
sample_size = 100
data = np.random.normal(mean, std_dev, sample_size)

cumulative_sum = np.cumsum(data)

# Plot cumulative sum
plt.plot(cumulative_sum)
plt.title('Cumulative Sum of Normal Distribution')
plt.xlabel('Index')
plt.ylabel('Cumulative Sum')
plt.grid(True)
plt.show()
```

```python
# Yes, the data  appears to follow an increasing trend from 0 to 100 and this increase is seen as consistent over the time indicating positive trend. 
```

```python
# 3. Choose 9 stocks of your choice and create a notebook with the same visualizations. 
# Provide similar explanations in regard to the meaning of statistics and the shape of each stocks distribution.


import yfinance as yf
import pandas as pd

car_brands = {
    'Toyota': 'TM',
    'Honda': 'HMC',
    'Tesla': 'TSLA',
    'BMW': 'BMWYY',
    'Kia': 'KIM',
    'Nissan': 'NSANY',
    'Ford': 'F',
    'General Motors': 'GM',
    'Stellantis': 'STLA'
}

start_date = '2023-01-01'
end_date = '2024-01-01'
stock_data = {}
for brand, ticker in car_brands.items():
    stock_data[brand] = yf.download(ticker, start=start_date, end=end_date)['Adj Close']

df = pd.DataFrame(stock_data)

df.plot(figsize=(12, 8), title='Stock Prices of Top 9 Car Brands in USA')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()
```
