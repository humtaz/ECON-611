```python
#  1. Create a list of random numbers between 0 and 100 whose length is 1000. (Hint: import random; search "python random" to learn more about the library.)

import random
random_numbers = [random.randint(0, 100) for i in range(1000)]
print(random_numbers)
```

```python
#  2. Use the variance function from the textbook to find the variance of this list. Assume that the list represent a population in whole.
import statistics
def variance(data, sample = True):
    mean = sum(data) / len(data)
    variance = sum((x - mean) ** 2 for x in data) / len(data)
    return variance
variance = variance(random_numbers)
print("Variance:", variance)
```

```python
#  3. Create a 9 more lists of the same length whose elements are random numbers between 0 and 100. Use a nested dictionary to house and identify these lists. 
# Keys for the first layer should be the numbers 1 through 10. Lists should be stored using a second key as follows: dict_name[index]["list"]. 
# Index represents the particular integer key between 1 and 10 as noted above.

lists_dict = {}
for i in range(1, 11):
    lists_dict [i] = {"list": [random.randint(0, 100) for _ in range(1000)]}
for key, value in lists_dict .items():
    print("Index:", key)
    print("List:", value["list"][:5])  
```

```python
# 5. At the end of chapter 2, we used for loops to find min and max values. 
# Create a min() function and max() function and pass the values from the list in question 1 to each of these to determine the min and max values in that list.

def find_min(data):
    return min(data)

def find_max(data):
    return max(data)

min_value = find_min(random_numbers)
max_value = find_max(random_numbers)

print("Minimum value:", min_value)
print("Maximum value:", max_value)
```
