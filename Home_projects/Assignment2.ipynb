```python
# 4. Import a data set for your project in pandas. You may use whatever means you prefer to import the data. Print the dataframe.
df_import = pd.read_csv('National_Obesity_By_State.csv')
df_import.head()
```

```python
# 5. Create a new column of data derived from the data that you have already imported. hint: this may be a ratio or whatever transformation that you think would be useful...
df_import['SHAPE_Length_km'] = df_import['SHAPE_Length'] / 1000
df_import['SHAPE_Area_km2'] = df_import['SHAPE_Area'] / (10**6)
df_import.head()
```

```python
# 6. Use a for loop to print individually each column in the dataframe.
for column in df_import.columns:
    print(df_import[column])
```

```python
# 7. Create a function that plots 2 to 3 different variables from the dataframe. The function may look something like "def plot_data(df,names, etc...). Each plot should include a legend and title.
import matplotlib.pyplot as plt
def plot_data(df, variables, x_label, y_label, title):
    plt.figure(figsize=(10, 6))
    for var in variables:
        plt.plot(df['FID'], df[var], label=var)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
# 8. Create a bar graph plot using the data you have imported.
plt.figure(figsize=(10, 6))
df_import['Obesity'].plot(kind='bar', color='skyblue')
plt.xlabel('FID')
plt.ylabel('Obesiety Data')
plt.title('Bar Graph Title')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# 9. On any of the graphs you have created, change the font size, axis labels, tick labels, color, linewidth, and so forth. Adjust these parameters to make the visualization aesthetically appealing..
# Increase font size 
plt.rc('font', size=12)
# figure and plot
plt.figure(figsize=(12, 8)) 
df_import['Obesity'].plot(kind='bar', color='skyblue', linewidth=2)
# axis labels and title font size 
plt.xlabel('FID', fontsize=14, fontweight='bold')
plt.ylabel('Obesity Data', fontsize=14, fontweight='bold')
plt.title('Bar Graph Title', fontsize=16, fontweight='bold')
# # Adjust tick label font size and weight
plt.xticks(rotation=45, fontsize=10, fontweight='bold')  
# gridlines
plt.grid(axis='y', linestyle='--', linewidth=0.5, color='gray')
plt.tight_layout()  
plt.show()
```
